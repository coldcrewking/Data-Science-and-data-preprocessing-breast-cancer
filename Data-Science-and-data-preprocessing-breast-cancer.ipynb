```python
import numpy as np
import tensorflow as tf
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'/Users/KHL/Downloads/breast_cancer.csv')
```

```python
X = dataset.iloc[:, 2:-1].values
```

```python
dataset['diagnosis'] = dataset['diagnosis'].replace({'M': 0,"B": 1})
```

```python
print("There Is No Value Are Duplicate:",dataset.duplicated().sum())
```

```python
dataset.describe()
```

```python
y = dataset.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
pd.set_option('display.float_format', lambda x: '%.4f' %x)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.layers import Dropout
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=28, activation='relu'))
ann.add(Dropout(0.2))
```

```python
ann.add(tf.keras.layers.Dense(units=26, activation='relu'))
ann.add(Dropout(0.2))
```

```python
ann.add(tf.keras.layers.Dense(units=26, activation='relu'))
ann.add(Dropout(0.2))
```

```python
ann.add(tf.keras.layers.Dense(units=26, activation='relu'))
ann.add(Dropout(0.2))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 300)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.0001)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
weights = ann.get_weights()
feature_importance_scores = np.mean(np.abs(weights[0]), axis=0)
```
